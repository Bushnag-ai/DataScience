# Project Proposal  - Corona(Covid-19)

## Stakeholders
This project aims to view the affects of a pandemic, in this case Covid-19, on the human lives and how each country's approch in handing the situation affected the survival,immunization,recovery, death..etc numbers 

The analysis would also show how the health-system, preparation and readiness: vaccine,ICU...etc, would affect the survival,immunization,recovery, death..etc numbers


This analysis would help future pandemics to be handled better, and can be used to detect negligence to human lives. Also which method was the most effective number wise and who would need the most help.

## Data Acquisition
The data was obtained from "Our World In Data" from this link https://ourworldindata.org/covid-deaths.
The data was downloaded as a CSV then converted to SQL.
The data before conversion was split into two CSV tables before conversion.

## Usage
The data will be used via SQL queries loaded into Pandas.DateFrame.
Once the data is in DataFrames the EDA process starts.
