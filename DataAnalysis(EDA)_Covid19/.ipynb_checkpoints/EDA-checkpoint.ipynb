# Data Analysis - EDA - Covid_19

This project aims to view the affects of a pandemic, in this case Covid-19, on the human lives and how each country's approch in handing the situation affected the survival,immunization,recovery, death..etc numbers

The analysis would also show how the health-system, preparation and readiness: vaccine,ICU...etc, would affect the survival,immunization,recovery, death..etc numbers

This analysis would help future pandemics to be handled better, and can be used to detect negligence to human lives. Also which method was the most effective number wise and who would need the most help.

First view of the data shows that the data has a column called index and the actual index column. This gives the reason to drop that column but the since the drop would only remove it from the dataframe and we will be using SQL to SELECT,join...etc from the database that tells that the drop would be a frequent. To make it easier a drop function of that row would be a good idea.

```python
def drop_extra(this_df, column_name = 'index', inplace = True):
    if inplace:
        this_df.drop(column_name, axis = True, inplace = inplace)
    else: 
        return this_df.drop(column_name, axis = True)
drop_extra(df)
columns_in_db.remove('index')
df.head()
```

#### Join Query
self-join

```python
print('self-join')

SELECT = "SELECT A.location as Country, A.population as populations, round(sum(B.icu_patients),0) as sum_icu_patients, round(sum(B.new_cases),0) as sum_cases, round(sum(B.new_deaths),0) as sum_deaths, round(sum(B.male_smokers),0) as Sum_Male_Smoker,round(sum(B.female_smokers),0) as Sum_Female_Smoker" 
FROM = f"FROM {table} A INNER JOIN {table} B"
ON = "ON A.date = B.date"
AND= "AND A.location = B.location"
GROUP_BY = "GROUP BY A.location, A.population"
ORDER_BY = "ORDER BY Country DESC"

query = f"{SELECT} {FROM} {ON} {AND} {GROUP_BY} {ORDER_BY};"
qdf = pd.read_sql(query,cnn)
qdf.head()
```

### Lets select the top 100 countries by number of deaths and sorting it
##### 1) lets see how is the number of deaths is affect by the increase in number of cases


```python
top100_pop = qdf.sort_values('sum_deaths', ascending = False).iloc[:100].reset_index(drop=True)
print(top100_pop.shape)
top100_pop.head(20)
```

##### Scatter plot to show the relationship between the increase of number of cases and number of deaths

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(style="ticks")
sns.scatterplot(data=top100_pop, x=top100_pop['sum_deaths'], y=top100_pop['sum_cases'], alpha=0.999,  size=top100_pop["sum_deaths"],
                sizes=(20, 200), hue = top100_pop["sum_deaths"])
plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plot.grid()  #just add this
plot.show()
```

#### To view the relationship better the outliars are removed in the next step. If the outliars are not affecting the trend then its most likely they habe no affect changing the general conclusion: like pulling towards a different conclusion. As we can see from the second plot below that the lower bounds of the data aligns with the outliars(the top 5)

```python
top100_pop = top100_pop[5:]
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(style="ticks")

sns.scatterplot(data=top100_pop, x=top100_pop['sum_deaths'], y=top100_pop['sum_cases'], alpha=0.999,  size=top100_pop["sum_deaths"],
                sizes=(20, 200), hue = top100_pop["sum_deaths"])

plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plot.grid()  
plot.show()
```

#### As we can see from the graph below when we remove 4 more outliars the relastionship continues. 

```python
top100_pop = top100_pop[15:]
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(style="ticks")

sns.scatterplot(data=top100_pop, x=top100_pop['sum_deaths'], y=top100_pop['sum_cases'], alpha=0.999,  size=top100_pop["sum_deaths"],
                sizes=(20, 200), hue = top100_pop["sum_deaths"])

plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plot.grid()  
plot.show()
```

##### After removing the outliars in stages we can observe if the outliars are affecting the relationship, which in this case they did not.
##### Then by looking at the scatter plot above we can conclude that the increase in number of cases did not always increase the number of death linearly. The Plot shows that some countries had the same number of cases as others but with higher deaths.
##### We can observe that where the number of cases is 0.5 which shows clearly a number of countries had the same number of cases but some had more casualties.
